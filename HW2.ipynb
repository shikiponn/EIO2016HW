```r
calc_LH <- function(df) {
    return(function(w){
        for(i in 1:nrow(df)){
            p1 = exp(X[,1] %*% w)
            p2 = exp(X[,2] %*% w)
            p3 = exp(X[,3] %*% w)
            p4 = exp(X[,4] %*% w)
            p5 = exp(X[,5] %*% w)
            p6 = exp(X[,6] %*% w)
            p7 = exp(X[,7] %*% w)
            p8 = exp(X[,8] %*% w)
            p9 = exp(X[,9] %*% w)
            p10 = exp(X[,10] %*% w)
            p0 = 
            sum = p1+p2+p3+p4+p5+p6+p7+p8+p9+p10+1
            df[i, "pchoice0"] = log(1-((sum-1)/sum))
            df[i, "pchoice1"] = log(p1/sum)
            df[i, "pchoice2"] = log(p2/sum)
            df[i, "pchoice3"] = log(p3/sum)
            df[i, "pchoice4"] = log(p4/sum)
            df[i, "pchoice5"] = log(p5/sum)
            df[i, "pchoice6"] = log(p6/sum)
            df[i, "pchoice7"] = log(p7/sum)
            df[i, "pchoice8"] = log(p8/sum)
            df[i, "pchoice9"] = log(p9/sum)
            df[i, "pchoice10"] = log(p10/sum)
        }   
        vec = as.matrix(df[,c("choice0","choice1","choice2","choice3","choice4","choice5","choice6","choice7","choice8","choice9","choice10")]*df[,c("pchoice0", "pchoice1","pchoice2","pchoice3","pchoice4","pchoice5","pchoice6","pchoice7","pchoice8","pchoice9","pchoice10")]) %*% as.matrix(numeric(11)+1)
#     print(head(df,3))
#         print( head(as.matrix(df[,c("choice0","choice1","choice2","choice3")]*df[,c("pchoice0", "pchoice1","pchoice2","pchoice3")])))
#         print(head(vec))
#         print(-sum(vec))
        #     print(head(vec,20))
    #  vec <- replace(vec, which(vec == 0), 1) 
    #  print(head(vec,20))
        #負の対数尤度を返す
        LH = -sum(vec)
        return(LH)
    })
}

```

```r
ns = 000#num of simulattion
ms = m*n
v = rnorm(ns*3)
v =matrix(v, ns, 3)



sigma = c(0.5, 0.5, 0.5)

calc_LH2 <- function(sigma){
    
    s0 = (ms-sum(df$market_share)/ms)
    sj = df$market_share / ms

    rep = t(matrix(c(numeric(ns)+1)))
    # print(dim(sigma))
    del0 = log(sj) - log(s0*(numeric(length(dim(df)[1]))+1))

    cont_map <- function(del) {
        Mat = rep %x%del
        n =t(t(rep %x% sigma) * v)
        X2 = as.matrix(X) %*% n
        Eu = exp(Mat + X2)
        denomi = matrix(c(numeric(ns)+1),1,ns)  + sum(Eu)
        denomimat = matrix(c(numeric(dim(df)[1])+1),dim(df)[1],1) %x% denomi
        probMat = Eu/denomimat
        SimMS = rowSums(probMat)/ns

        init_del=matrix(del)
        updated_del = matrix(init_del + log(sj) - log(SimMS))
        diffd= sum((init_del - updated_del))
        return(list(diffd, updated_del))
    }
    
    del = del0
    diffd =1
    while (diffd  >0.000001) {
        res =    cont_map(del)
        diffd = res[1]
        del =res[2]
    }


    X = data.frame(X, del)
    names(X) <- c("const", "x1", "x2", "del")
    res = lm(del ~ const + x1+x2, data = X)

    print(summary(res))
    xi = res$residuals
    # print("Xi is")
    # print(head(xi))
    W = diag(length(xi)) #重み
    min  = t(xi) %*% W %*% xi
    return(min)

}

res = optim(par = c(0.5, 0.5, 0.5), fn = calc_LH2)
# for(i in 1:ns){}
```
